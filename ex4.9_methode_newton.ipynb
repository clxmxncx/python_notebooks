Résolution de l'équation $x^3 - 2x^2 + 1$ avec la méthode de Newton puis affichage des résultats 

<br>


#### question 0.
Implementer la méthode de Newton avec comme critère d'arrêt $\lvert f(x) \rvert < \epsilon$

#### question 1.
Résoudre l'équation précédente pour $x_0$ qui varie entre $-1$ et $3$ par pas de $0.01$
et représenter graphiquement la valeur de la solution trouvée en fonction de $x_0$.

```python
import numpy as np
from matplotlib import pyplot as plt

def f(x):
    return x**3 - 2*x**2 + 1

def f_prime(x):
    return 3*x**2 - 4*x

def newton(f,f_prime,x0,eps):
    x = x0
    while np.abs(f(x)) >= eps:
        y = f(x)
#         print("x : ",x)
#         print("y : ",y)
        y_prime = f_prime(x)
#         print("y_prime : ", y_prime)
        if np.abs(y_prime)>0.0000001:
            x = x-(y/y_prime)
        else:
            return None
    return x


# https://matplotlib.org/3.1.0/gallery/subplots_axes_and_figures/subplots_demo.html
# https://matplotlib.org/tutorials/intermediate/gridspec.html
fig = plt.figure(figsize=(10, 8), constrained_layout=False)
gs1 = fig.add_gridspec(nrows=3, ncols=1, left=0.05, right=0.48, hspace=0.5)
ax1 = fig.add_subplot(gs1[:-1, :])

x = np.linspace(-1, 3, 100)
y = f(x)
z = [0]*(len(x))    
ax1.set_ylim(-1, 10)
ax1.plot(x, y, label='$y=x^3 - 2x^2 + 1$')
ax1.plot(x, z, label='$y=0$')
ax1.legend()

ax2 = fig.add_subplot(gs1[-1, :])

x = []
sol = []
scale = 1000
for i in range(-1 * scale, 3 *scale):
    x.append(i/scale)
    t = newton(f, f_prime, i/scale, 1e-6)
#     print(t)
    sol.append(t)
ax2.scatter(x, sol, s=2)
# ax2.plot(x, sol)


plt.show()
```

#### Explication de la cellule ci-dessous

Si $x_0 = 0$, $f'(x_0) = 0$ et donc la tangente ne coupe pas 
l'axe des abscisses. La suite ne peut pas être construite
et le programme renvoie `None`.

<br>


```python
eps = 1e-6
res = newton(f, f_prime, 0, eps)
print(res)
```
